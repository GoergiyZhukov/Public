# Задание
Изобразить графически как за последние 10 лет изменилась численность населения Дубая и предложить свой вариант прогноза, используя наиболее подходящий на ваш взгляд подход

Данные по численности населения вяты с сайта https://ru.zhujiworld.com/ae/133170-dubai/

```python
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
from scipy import stats
import seaborn as sns
import warnings
warnings.simplefilter(action='ignore')

sns.set(style='whitegrid', rc={'figure.figsize' : (18,8)})
```

```python
df = pd.read_excel('Populatin_Dubai.xlsx')
```

```python
df
```

Для визуализации изменения численности населения решил использовать обычный линейный график из библиотек seaborn и диаграмму рассеяния из библиотек matplotlib

```python
sns.lineplot(data=df, x="year", y="population")
plt.show()
```

```python
plt.scatter (df.year, df.population)
plt.title('year vs. population')
plt.xlabel('year')
plt.ylabel('population')
plt.show()
```

Из графиков видно, что прирост населения шёл практически линейно. Выбросов на графике нет.

Т.к. дополнительных параметров оценки не задано и в условиях одна переменная независимая (year), а вторая зависимая (population), считаю сделать прогноз увеличения численности населения на основе линейной регресси.

```python
sns.set(style='whitegrid', rc={'figure.figsize' : (20,8)})
sns.regplot(x='year', y='population', data=df)
plt.title('Взаимосвязь изменения численности населения Дубая')
plt.ylabel('Население, млн')
plt.xlabel('Год');
```

Если принебречь другими факторами роста численности населения (экономические, политические, социальные ит.д.), а также исключить "чёрных лебедей", то динамика роста будет выглядеть как указано на графике выше. Единственным недостатком будет являться небольшой объём данных, поэтому для большей достоверности результатов лучше взять более длительный период. Для этого имеются данные с 50х годов.
