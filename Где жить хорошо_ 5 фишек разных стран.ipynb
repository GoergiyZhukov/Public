### Небольшие умозаключения по поводу, где Ивану жить хорошо (или не очень)

1. Швейцарию и Норвегию по достоинству смогут оценить состоятельные граждане, которые предпочитают горнолыжный отдых и не любят переплачивать за жильё: при достаточно высоких показателях основных индексов имеют относительно невысокий Индекс аренды.
2. Для тех, кто любит высокий уровень жизни и не любит готовить и переплачивать за жильё, отлично подойдёт Южная Корея или Япония: низкие Индексы аренды и цен ресторанов при высоком Индексе стоимости жизни и достаточно большом Продуктовом индексе.
3. Если вы любите небольшие южные страны, где всего в меру + наличие морских пространств неподалёку, а стоимость аренды Вас не сильно смущает, то Вам отлично подойдут Гонконг или Сингапур: достаточно усреднённые показатели основных индексов и высокий Индекс аренды.
4. Любите недорогую европейскую глубинку - добро пожаловать в Косово: низкий уровень Индекса аренды и очень хорошая Местная покупательская способность.
5. Индия - то же Косово, только в профиль, т.е. с морем, солнцем и дождём. Если любите южные страны и не любите расстатваться с шекелями.
