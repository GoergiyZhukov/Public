```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as mp
```

```python
df = pd.read_csv('Cost_of_Living_Index_2022.csv')
```

```python
df.head()
```

```python
df.rename(columns={'Unnamed: 0':'rank', 'Unnamed: 1': 'country ',
                       'Unnamed: 2': 'cost_of_living_index', 'Unnamed: 3': 'rent_index',
                       'Unnamed: 4': 'cost_of_living_plus_index', 'Unnamed: 5': 'groceries_index',
                       'Unnamed: 6': 'restaurant_price_index', 'Unnamed: 7': 'local_purchasing_power_index'
                        }, inplace=True)
```

```python
df.head()
```

```python
df = df.drop(index=0)
```

```python
df.shape
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
df.describe()
```

```python
df.dtypes
```

```python
df[['cost_of_living_index', 'rent_index', 'cost_of_living_plus_index', 'groceries_index',
    'restaurant_price_index', 'local_purchasing_power_index']] = df[['cost_of_living_index', 'rent_index',
                                                                     'cost_of_living_plus_index', 'groceries_index', 
      'restaurant_price_index', 'local_purchasing_power_index']].astype (str).astype (float)
```

```python
df.dtypes
```

```python
df.sort_values(by=['cost_of_living_index', 'local_purchasing_power_index'], ascending=False).head(20)
```

```python
top20 = df.sort_values(by=['cost_of_living_index', 'local_purchasing_power_index'], ascending=False).head(20)
```

```python
top20.plot(x="rank", y=['cost_of_living_index', 'rent_index', 'cost_of_living_plus_index', 
                        'groceries_index', 'restaurant_price_index', 'local_purchasing_power_index'], 
           kind="bar", figsize=(20, 8))
```

```python
df.sort_values(by=['cost_of_living_index', 'local_purchasing_power_index'], ascending=False).tail(20)
```

```python
end20 = df.sort_values(by=['cost_of_living_index', 'local_purchasing_power_index'], ascending=False).tail(20)
```

```python
end20.plot(x="rank", y=['cost_of_living_index', 'rent_index', 'cost_of_living_plus_index', 
                        'groceries_index', 'restaurant_price_index', 'local_purchasing_power_index'], 
           kind="bar", figsize=(20, 8))
```
